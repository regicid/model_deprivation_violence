# Figure S2 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
PI = [20,30,40,50]


min_ = .0001
max_ = .003
def desp_rate(sigma=6,mu=15):
    return norm.cdf(0,loc=mu,scale=sigma)
MU = np.linspace(8,30,10000)
desp_rates = [desp_rate(mu=mu) for mu in MU]
desp = np.linspace(min_,max_,50)
MU = [MU[np.argmin(np.abs(x-desp_rates))] for x in desp]
initial_v=1

for pi in PI:
    results = [np.load(f"Results/result2_{np.round(mu,2)}_pi={pi}_v={initial_v}.npy") for mu in MU]
    V = [1-result[0,-1] for result in results]
    plt.scatter(desp,V,label=f"π = {pi}")
plt.xticks([0,.001,.002,.003])
plt.legend()
plt.xlabel("Desperation rate")
plt.ylabel("Violence at equilibrium")
plt.savefig("Supplementary_figures/sensitivity_pi.jpg",dpi=300,bbox_inches="tight")
```
